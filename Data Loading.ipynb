## Data Inputs

```python
#######################
#### DATA INPUTS ######
#######################

exchange_rate_files = ['eurgbp_2015_2020.csv', 'eurjpy_2015_2020.csv', 'eurusd_2015_2020.csv', 'gbpjpy_2015_2020.csv','gbpusd_2015_2020.csv']
index_files = ['GSPC_2015_2020.csv', 'VIX_2015_2020.csv']
```

## Extract and Clean

```python
# Extracting data
exchange_rate_data = {file[:6] : pd.read_csv(file, index_col='date', parse_dates=True) for file in exchange_rate_files}
index_data = {file.split('_')[0].lower() : pd.read_csv(file, index_col='Date', parse_dates=True) for file in index_files}
print(f'Exchange Rate Keys: \t {list(exchange_rate_data.keys())} \n' +
        f'Index Keys: \t \t {list(index_data.keys())}')
```

```python
## Looking at what data is null
test = exchange_rate_data['eurgbp'].join(index_data['gspc'])
print(f'Number of null values is {len(np.sum(test.isnull()))} out of {len(test)}')
```

```python
sns.heatmap(test.isnull())
```

## Visualize

```python
for df in exchange_rate_data.values():
    print(len(df), np.sum(df.isnull()))
    df.plot()
    
for df in index_data.values():
    print(len(df), np.sum(df.isnull()))
    df.plot()
```
